# Data Cleaning

```python
# 'Neighborhood'
# replace duplicates: Pan Handle with Panhandle

housing['Neighborhood'].replace(
    to_replace=['Panhandle'],
    value='Pan Handle',
    inplace=True
)
housing['Neighborhood'].replace(
    to_replace=['Telegraph hill'],
    value='Telegraph Hill',
    inplace=True
)
housing['Neighborhood'].replace(
    to_replace=['duboce triangle','Duboce triangle'],
    value='Duboce Triangle',
    inplace=True
)
housing['Neighborhood'].replace(
    to_replace=['Nob hill'],
    value='Nob Hill',
    inplace=True
)

housing['Neighborhood'].replace(
    to_replace=['bernal Heights'],
    value='Bernal Heights',
    inplace=True
)
housing['Neighborhood'].replace(
    to_replace=['portrero hill'],
    value='Portrero Hill',
    inplace=True
)
housing['Neighborhood'].replace(
    to_replace=['Inner sunset'],
    value='Inner Sunset',
    inplace=True
)
housing['Neighborhood'].replace(
    to_replace=['Outer sunset'],
    value='Outer Sunset',
    inplace=True
)
```

```python
# 'Condition and Quality':

# 1.find out null values
# in this case, only numeric value and null exist
print(housing['Condition and Quality'].isnull())
```

```python
# 2.we are going to replace null value with median - 5.00 based on right-skewness.

housing['Condition and Quality'].fillna(housing['Condition and Quality'].median(),inplace=True)
```

```python
# Walkability
# False value

housing['Walkability Score'].replace(
    to_replace=['0','False'],
    value=84,
    inplace=True
)
print(housing.iloc[2303,:])
```

```python
# '2 stories?'
# since outliers are only a few data entries, 
# we can interpreting it as 'no' in this model


housing['2 stories?'].replace(
    to_replace=['3', '3 stories','4 stories'],
    value='no',
    inplace=True
)

# change to boolean value 1 and 0

housing['2 stories?'].replace(
    to_replace=['yes'],
    value=1,
    inplace=True
)
housing['2 stories?'].replace(
    to_replace=['no'],
    value=0,
    inplace=True
)
```

```python
#'Lot Location' 
# we predict FR3 based on price sold can be seen as Corner

housing['Lot Location'].replace(
    to_replace=['FR3'],
    value='Corner',
    inplace=True
)
```

```python
# Lot Location
"""housing['Lot Location'].replace(
    to_replace=['Inside'],
    value=obj_houses,
    inplace=True
)"""
"""housing['Lot Location'].replace(
    to_replace=['Corner'],
    value=2,
    inplace=True
)
housing['Lot Location'].replace(
    to_replace=['CulDSac'],
    value=3,
    inplace=True
)"""
```

```python
# 'Sq ft living space'
# since it is right-skewed, we are replacing 'Not avail' to the median 1579:

housing['Sq ft living space'].replace(
    to_replace='Not avail',
    value= 1579,
    inplace=True
)
```

```python
# fill in any NA value with 2 as predicted
housing["No. parking spaces"].fillna(value=2, inplace = True)
```
